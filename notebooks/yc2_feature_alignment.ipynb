```python
import pandas as pd
import os
import json
import numpy as np
```

```python
d_files = os.listdir('raw_videos/')
d_ids = [x[8:19] for x in d_files]
d_ids
```

```python
ROOT = '/network/scratch/a/aasheesh.singh/YouCookII/features/feat_csv/'

paths = []

for part in os.listdir(ROOT):
    for id in os.listdir(ROOT + '/' + part):
        for yt_id in os.listdir(ROOT + '/' + part + '/' + id):
            if yt_id in d_ids:
                paths.append(ROOT + '/' + part + '/' + id + '/' + yt_id)

# os.listdir('train_frame_feat_csv/227/p8ZGmL-8w6E/0010')

paths
```

```python
paths_with_ten_segs = []

for path in paths:
    if len(os.listdir(path)) == 10:
        paths_with_ten_segs.append(path)


len(paths_with_ten_segs)
# set(paths) - set(paths_with_ten_segs)
```

```python
for path in paths_with_ten_segs:
    data = []

    if not os.path.exists('features/' + path[-11:]):
        os.mkdir('features/' + path[-11:])

    segs = os.listdir(path)
    segs.sort()
    for seg in segs:
        df = pd.read_csv(path + '/' + seg + '/resnet_34_feat_mscoco.csv', header = None)
        data.append(df)
    
    final_df = pd.concat(data, ignore_index = True)
    final_df.to_csv('features/' + path[-11:] + '/master_features.csv', header = False)
    print(path[-11:], 'is complete!')
```

```python
with open('youcookii_annotations_trainval.json') as f:
    train_val_json = json.load(f)

with open('youcookii_annotations_test_segments_only.json') as f:
    test_json = json.load(f)


for yt_id in os.listdir('features'):
    duration = 0
    
    if yt_id in train_val_json['database'].keys():
        duration = train_val_json['database'][yt_id]['duration']
    elif yt_id in test_json['database'].keys():
        duration = test_json['database'][yt_id]['duration']
    else:
        print(yt_id, "not found!")

    
    clip_duration = duration / 10.0     # Total duration divided into 10 equal clips
    fps = 500.0 / clip_duration         # Assumption: each clip contributes exactly 500 frames to the total 5000 features
    unit_time = 1.0 / fps               # Unit time of each frame

    timestamps = np.linspace(0, unit_time*5000, 5001)[:-1]      # Selecting the first 5000 elements
    timestamps = np.around(timestamps, 2)


    df = pd.read_csv('features/' + yt_id + '/master_features.csv', header = None) 
    df['timestamps'] = timestamps
    df.to_csv('features/' + yt_id + '/master_features_with_ts.csv', header = False)
```
