```python
# zBNvyUEvjtY
ROOT = '/network/scratch/a/aasheesh.singh/YouCookII/splits/'

test_df = 0

for yt_id in os.listdir('features'):
    print(yt_id)

    df_root_train = pd.read_csv(ROOT + 'train_duration_totalframe.csv')
    df_root_val = pd.read_csv(ROOT + 'val_duration_totalframe.csv')
    df_root_test = pd.read_csv(ROOT + 'test_duration_totalframe.csv')

    total_frames, duration = 0, 0 
    
    if yt_id in df_root_train.vid_id.values:
        total_frames = df_root_train[df_root_train.vid_id == yt_id]['total_frame'].values[0]
        duration = df_root_train[df_root_train.vid_id == yt_id]['duration'].values[0]
        print(total_frames, duration)
    elif yt_id in df_root_val.vid_id.values:
        total_frames = df_root_val[df_root_val.vid_id == yt_id]['total_frame'].values[0]
        duration = df_root_val[df_root_val.vid_id == yt_id]['duration'].values[0]
        print(total_frames, duration)
    else:
        total_frames = df_root_test[df_root_test.vid_id == yt_id]['total_frame'].values[0]
        duration = df_root_test[df_root_test.vid_id == yt_id]['duration'].values[0]
        print(total_frames, duration)

    fps = total_frames / duration
    unit_time = 1.0 / fps

    sampling_interval = total_frames / 500
    segs = 10

    print(fps, sampling_interval)

    times = []
    for i in range(segs):
        if i==0:
            times.append(0.0)
            start_ts = 0.0
        else:
            start_ts = np.max([np.floor(sampling_interval/segs) , 1]) * i * unit_time
            times.append(start_ts)

    data = gen_series(times)
    df_ts = pd.concat(data).sort_index(kind='stable').reset_index(drop = True)

    df = pd.read_csv('features/' + yt_id + '/master_features_v2.csv', header = None)
    df['timestamp'] = df_ts[0]

    df.to_csv('features/' + yt_id + '/master_features_with_ts_v2.csv', header = False)
    print(yt_id, 'is complete!')




# test_df.tail()
    
```
